```python
file_path = os.getcwd()
file_name = Path(file_path, "USW00094973.csv")
file_name
```

```python
# first inpsect at data. see how it looks like
df = pd.read_csv(file_name)  
df
```

```python
def get_hotest_days(file_name = file_name, drop_or_interpolate = "drop"):
    """
        input: us temparature and prec historical data
        output: hostest day
        notes: drop_or_interpolate only accept two values : drop or interpolate
            if drop_or_interpolate set to drop, will drop all rows with missing data
            if interpolate: fill with neigboring data. first forward, then backward        
    """    
    df = pd.read_csv(file_name)  
    if not drop_or_interpolate:
        print("please specify missing data handling decision")
        return
    if drop_or_interpolate.lower() not in ["drop", "interpolate"]:
        print("unknown missing data handling decision")
    
    if drop_or_interpolate == "drop":
        df.dropna(inplace = True)
    else: 
        df = df.interpolate(method ='linear', limit_direction ='forward')
        df = df.interpolate(method ='linear', limit_direction ='backward')
    
    # the data is row by row. we either use build-in idxmax (similar to argmax) to 
    # get the row id with the hotest day
    # this method requires you be very familiar with pands the following:
    """
    row = df.loc[df['tmax'].idxmax()] 
    return row[1], row[2] 
    """ 
    # or we loop the all rows from begining to the end and find the hotest one
    # to practice
    row_id_with_max, max_temp = 0, -99
    for i, row in enumerate(df.values):
        if row[2] > max_temp: 
            max_temp = row[2]
            row_id_with_max = i
    row = df.loc[row_id_with_max]  
    return row[1],  max_temp 
    
     
```
